### Reduction operation: the sum of the numbers in the range [0, a)

Por otro lado, la función sum() de Python mejora considerablemente el tiempo de ejecución gracias a su implementación optimizada en C, permitiendo una suma mucho más rápida de los elementos de la lista. Finalmente, np.sum() de NumPy ofrece un rendimiento similar al de sum(), pero su verdadera ventaja radica en su capacidad para manejar arrays más grandes y multidimensionales de forma más eficiente, aprovechando las optimizaciones propias de NumPy. Esto demuestra que usar funciones nativas optimizadas o bibliotecas especializadas como NumPy puede acelerar significativamente el procesamiento en comparación con enfoques más simples como los bucles for.

```python
from numba import njit
import numpy as np
import time

# Versión original con NumPy (para comparar)
def numpy_sum_original(arr):
    return np.sum(arr)

# Versiones optimizadas con Numba
@njit
def numpy_sum_numba_optimized(arr):
    """Suma con np.sum() + Numba"""
    return np.sum(arr)

@njit
def manual_sum_numba(arr):
    """Suma con bucle for + Numba"""
    x = 0
    for i in arr:
        x += i
    return x

# Configuración del array (optimizado para Numba)
value = 10**6
array = np.arange(value, dtype=np.int64)  # dtype explícito para mejor rendimiento

# ---- Comparativa de tiempos ----
print("=== Comparativa NumPy vs. Numba (1 millón de elementos) ===")

# 1. NumPy puro
start = time.time()
result_numpy = numpy_sum_original(array)
numpy_time = time.time() - start
print(f"\nNumPy puro: {numpy_time:.6f} segundos")

# 2. Numba + np.sum() (incluyendo compilación JIT)
start = time.time()
result_numba_sum = numpy_sum_numba_optimized(array)  # Primera ejecución (compila)
numba_sum_time = time.time() - start
print(f"\nNumba + np.sum() - Primera ejecución: {numba_sum_time:.6f} segundos")

start = time.time()
result_numba_sum = numpy_sum_numba_optimized(array)  # Segunda ejecución (ya compilado)
numba_sum_time_compiled = time.time() - start
print(f"Numba + np.sum() - Segunda ejecución: {numba_sum_time_compiled:.6f} segundos")

# 3. Numba + bucle manual
start = time.time()
result_manual_numba = manual_sum_numba(array)  # Primera ejecución (compila)
numba_manual_time = time.time() - start
print(f"\nNumba + bucle for - Primera ejecución: {numba_manual_time:.6f} segundos")

start = time.time()
result_manual_numba = manual_sum_numba(array)  # Segunda ejecución (ya compilado)
numba_manual_time_compiled = time.time() - start
print(f"Numba + bucle for - Segunda ejecución: {numba_manual_time_compiled:.6f} segundos")

# ---- Resultados ----
print("\n=== Resumen de resultados ===")
print(f"Suma NumPy puro: {result_numpy}")
print(f"Suma Numba + np.sum(): {result_numba_sum}")
print(f"Suma Numba + bucle for: {result_manual_numba}")

# Speedup comparativo
speedup_sum = numpy_time / numba_sum_time_compiled
speedup_manual = numpy_time / numba_manual_time_compiled
print(f"\nSpeedup Numba+np.sum() vs NumPy: {speedup_sum:.1f}x")
print(f"Speedup Numba+bucle vs NumPy: {speedup_manual:.1f}x")
```
